# Simulation 1

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

VecbS = [4,8,12,16,20,24]
```

```python
%run sim001.py
```

#### Front 95/naive/zeros comparison

```python
plt.rcParams['axes.spines.right'] = False
plt.rcParams['axes.spines.top'] = False

plt.figure(figsize=(24, 12))

fig,ax = plt.subplots()

plt.errorbar(VecbS, mean_front_95, yerr=sd_front_95, fmt='-', color='blue', label=r'$q_{95}$')
plt.errorbar(VecbS, mean_front_Naive, yerr=sd_front_Naive, fmt='-', color='red', label=r'Medida sencilla')
plt.errorbar(VecbS, mean_front_zeros, yerr=sd_front_zeros, fmt='-', color='green', label=r'Primer cero')
plt.xticks(VecbS)
ax.set(xlabel=r'Número de canales ($b_{S}=b_{T}$)', ylabel='Posición del frente ($F$)',
       title='')
plt.grid(axis='y')
plt.legend(loc='best')
plt.savefig('Motilities_bs_bt.png', dpi=1080)
plt.show()
```

#### Front 95 comparison

```python
plt.figure(figsize=(24, 12), dpi=80)

fig,ax = plt.subplots()

plt.errorbar(VecbS, mean_front_95, yerr=sd_front_95, fmt='-', color='green', label=r'$q_{95}$')
plt.errorbar(VecbS, mean_front_95CSC, yerr=sd_front_95CSC, fmt='-', color='blue', label =r'CSC')
plt.errorbar(VecbS, mean_front_95TD, yerr=sd_front_95TD, fmt='-', color='red', label =r'TD')
plt.xticks(VecbS)
ax.set(xlabel=r'Number of channels ($b_{S}=b_{T}$)', ylabel='Front position ($F$)',
       title='Motilities front 95')
plt.grid(axis='y')
plt.legend(loc='upper right')
plt.savefig('Motilities_95comparison.png', dpi=1080)
plt.show()
```

```python
plt.figure(figsize=(24, 12), dpi=80)

fig,ax = plt.subplots()

plt.plot(VecbS,
         np.abs(np.array(mean_front_95CSC)-np.array(mean_front_95TD)),
         linestyle='-', color='Orange'
        )
plt.xticks(VecbS)
ax.set(xlabel=r'Número de canales ($b_{S}=b_{T}$)', ylabel='Diferencia',
       title='Diferencia en valor absoluto del $q_{95}$ de CSC y TD')
plt.grid(axis='y')
plt.savefig('Motilities_95comparisonDiference.png', dpi=1080)
plt.show()
```

#### Front Naive Comparison

```python
plt.figure(figsize=(24, 12), dpi=80)

fig,ax = plt.subplots()

plt.errorbar(VecbS, mean_front_Naive, yerr=sd_front_Naive, fmt='-', color='green', label=r'Naive measure')
plt.errorbar(VecbS, mean_front_NaiveCSC, yerr=sd_front_NaiveCSC, fmt='-', color='blue', label =r'CSC')
plt.errorbar(VecbS, mean_front_NaiveTD, yerr=sd_front_NaiveTD, fmt='-', color='red', label =r'TD')
plt.xticks(VecbS)
ax.set(xlabel=r'Number of channels ($b_{S}=b_{T}$)', ylabel='Front position ($F$)',
       title='Motilities front Naive')
plt.grid(axis='y')
plt.legend(loc='upper right')
plt.savefig('Motilities_Naivecomparison.png', dpi=1080)
plt.show()
```

```python
plt.figure(figsize=(24, 12), dpi=80)

fig,ax = plt.subplots()

plt.plot(VecbS,
         np.abs(np.array(mean_front_NaiveCSC)-np.array(mean_front_NaiveTD)),
         linestyle='-', color='Orange'
        )
plt.xticks(VecbS)
ax.set(xlabel=r'Number of channels ($b_{S}=b_{T}$)', ylabel='Difference',
       title='Absolute value of difference between CSC and TD')
plt.grid(axis='y')
plt.savefig('Motilities_NaivecomparisonDiference.png', dpi=1080)
plt.show()
```

#### Front zeros comparison

```python
plt.figure(figsize=(24, 12), dpi=80)

fig,ax = plt.subplots()

plt.errorbar(VecbS, mean_front_zeros, yerr=sd_front_zeros, fmt='-', color='green', label=r'First zero')
plt.errorbar(VecbS, mean_front_zerosCSC, yerr=sd_front_zerosCSC, fmt='-', color='blue', label=r'CSC')
plt.errorbar(VecbS, mean_front_zerosTD, yerr=sd_front_zerosTD, fmt='-', color='red', label=r'TD')
plt.xticks(VecbS)
ax.set(xlabel=r'Number of channels ($b_{S}=b_{T}$)', ylabel='Front position ($F$)',
       title='Motilities front Zeros')
plt.grid(axis='y')
plt.legend(loc='upper right')
plt.savefig('Motilities_zeroComparison.png', dpi=1080)
plt.show()
```

```python
plt.figure(figsize=(24, 12), dpi=80)

fig,ax = plt.subplots()

plt.plot(VecbS,
         np.abs(np.array(mean_front_zerosCSC)-np.array(mean_front_zerosTD)),
         linestyle='-', color='Orange'
        )
plt.xticks(VecbS)
ax.set(xlabel=r'Number of channels ($b_{S}=b_{T}$)', ylabel='Difference',
       title='Absolute value of difference between CSC and TD')
plt.grid(axis='y')
plt.yscale('log')
plt.savefig('Motilities_ZeroscomparisonDiference.png', dpi=1080)
plt.show()
```

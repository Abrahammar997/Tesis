```python
alpha = [0,1,2,3,4]
```

```python
plt.rcParams['axes.spines.right'] = False
plt.rcParams['axes.spines.top'] = False

plt.figure(figsize=(24, 12), dpi=200)

fig,ax = plt.subplots()

plt.errorbar(alpha, mean_front_95, yerr=sd_front_95, fmt='-', color='blue', label=r'$q_{95}$')
plt.errorbar(alpha, mean_front_Naive, yerr=sd_front_Naive, fmt='-', color='red', label=r'Medida sencilla')
plt.errorbar(alpha, mean_front_zeros, yerr=sd_front_zeros, fmt='-', color='green', label=r'Primer cero')
plt.xticks(alpha)
ax.set(xlabel=r'Fuerza de retroalimentación ($\alpha$)', ylabel='Posición del frente ($F$)',
       title='')
plt.grid(axis='y')
lg = plt.legend(bbox_to_anchor=(1,0.5), loc="center left")
plt.savefig('Motilities_alpha.png', 
            dpi=1080, 
            bbox_extra_artists=(lg,),
            bbox_inches='tight'
           )
plt.show()
```

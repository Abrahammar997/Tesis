```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

from subprocess import call
import os

from matplotlib import animation
from IPython.display import HTML
plt.rcParams['figure.dpi'] = 80

from scipy.integrate import odeint
from scipy import signal
```

```python
altura = 100
bS = 8
bT = 4
nu = 0.8
alpha = 0
psym = 0.3
death = 0.05
time = 300

os.remove('ydata.csv')
os.remove('ydataS.csv')
os.remove('ydataT.csv')
call(['./sim004.out',str(altura),str(bS),str(bT),str(nu),str(alpha),str(psym),
      str(death),str(time)])

ydata = pd.read_csv('ydata.csv',delimiter=' ',header=None)
ydata.drop(ydata.columns[-1], axis=1, inplace=True)

ydataS = pd.read_csv('ydataS.csv',delimiter=' ',header=None)
ydataS.drop(ydataS.columns[-1], axis=1, inplace=True)

ydataT = pd.read_csv('ydataT.csv',delimiter=' ',header=None)
ydataT.drop(ydataT.columns[-1], axis=1, inplace=True)

ylim = max(ydataT.values.max(),ydataS.values.max())

fig, ax = plt.subplots()
ax.set(xlabel='Column', ylabel='Cells',
       title=r'Y-density ($b_S$='+str(bS)+', $b_T$='+str(bT)+' )', xlim=[0,100],ylim=[0,ylim])
ax.grid()

line1, = ax.plot([],[],'b-',label="CSC")
line2, = ax.plot([],[],'r-',label="TD")

plt.legend(loc="upper right")

def drawframe(n):
    line1.set_data(range(100),ydataS.iloc[n,0:100])
    line2.set_data(range(100),ydataT.iloc[n,0:100])
    return (line1,line2)
anim = animation.FuncAnimation(fig, drawframe,
                               frames=range(int(time)),blit=True)
anim.save('Y_density_'+str(bS)+'_b_T_'+str(bT)+'_alpha_'+str(alpha)+'.gif', fps=60)


```

```python
HTML(anim.to_html5_video())
```
